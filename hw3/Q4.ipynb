```python
#Max Shi CS 559 Homework 3 Question 4
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
```

```python
train_data = pd.read_csv(r"E:\Repositories\cs559\hw3\train.csv")
test_data = pd.read_csv(r"E:\Repositories\cs559\hw3\test.csv")
```

```python
def distance_weight(distances):
    result = list( map(lambda three: list(map(lambda x: 1/(x*x), three)), distances))
    return result
#function for inverse square of distances for weights
```

```python
ThreeNearest = KNeighborsClassifier(n_neighbors=3, weights="uniform")
EuclideanDistance = KNeighborsClassifier(n_neighbors=3, weights=distance_weight, metric="euclidean")
```

```python
train_x = train_data[train_data.columns[:-1]]
train_y = train_data.iloc[: , -1]
print(train_x)
print(train_y)
```

```python
ThreeNearest.fit(train_x, train_y)
EuclideanDistance.fit(train_x, train_y)
#fit models
```

```python
test_x = test_data[test_data.columns[1:-1]]
ThreeNearest_predictions = ThreeNearest.predict(test_x)
print(ThreeNearest.predict_proba(test_x)) #print out probabilty estimates
EuclideanDistance_predictions = EuclideanDistance.predict(test_x)
```

```python
test_results = test_data.copy()
test_results["three-nearest-predict"] = ThreeNearest_predictions.tolist()
test_results["euclid-distance-predict"] = EuclideanDistance_predictions.tolist()
test_results #list out predictions from the two models
```

```python
def test_data_error(Y, pred_val):
    errors = 0
    TP = 0
    FP = 0
    FN = 0
    TN = 0
    for x in range(len(pred_val)):
        if(pred_val[x] == 1 and pred_val[x] == Y.iloc[x]):
            TP += 1
        elif(pred_val[x] == 0 and pred_val[x] == Y.iloc[x]):
            TN += 1
        elif(pred_val[x] == 1 and pred_val[x] != Y.iloc[x]):
            FP += 1
        elif(pred_val[x] == 0 and pred_val[x] != Y.iloc[x]):
            FN += 1
    #print(errors, 1-(errors / len(pred_val)))
    accuracy = (TP + TN) / (TP + TN + FP + FN)
    precision = TP / (TP + FP)
    recall = TP / (TP + FN)
    fmeasure = 2*precision*recall / (precision+recall)
    print("              Predicted Value")
    print("                 1       0   ")
    print("Actual    1:   %3d     %3d   " %(TP, FN))
    print("          0:   %3d     %3d   " %(FP, TN))
    results = {"accuracy": accuracy, "precision": precision, "f-measure": fmeasure}
    print(results)
```

```python
three_nearest_results = test_data_error(test_results["actual-class"], test_results["three-nearest-predict"])
#print out confusion matrix, accuracy, precision, and f-measure
```

```python
euclid_results = test_data_error(test_results["actual-class"], test_results["euclid-distance-predict"])
#print out confusion matrix, accuracy, precision, and f-measure
```

```python
labels = raw_data.loc[0:,2:]
rge = labels.max() - labels.min()
labels = (labels - labels.min())/rge
labels = pd.DataFrame(np.hstack((np.matrix(np.ones(labels.shape[0])).T, labels)))
labels
```

```python
classification = raw_data.loc[0:, 1]
classification = classification.apply(lambda x: 1 if x=='M' else 0)
classification
```

```python
def sigmoid_func(w,X):
    return 1/(1+np.exp(-np.dot(X,w.T)))
w = np.matrix(np.zeros(labels.shape[1]))
w1 = pd.DataFrame(sigmoid_func(w, labels))
```

```python
def cost_func(w,X,Y):
    func_result = sigmoid_func(w,X)
    #print(np.log(func_result).transpose()*Y)
    print(Y)
    first = (-np.log(func_result).transpose()*Y).transpose()
    second = (-np.log(1-func_result).transpose()*(1-Y)).transpose()
    final = first + second
    return final.mean()
```

```python
def error_function_deriv(w,X,Y):
    sigmoid = pd.DataFrame(sigmoid_func(w,X))
    #print("sigmoid", sigmoid)
    #print("Y", Y)
    first = sigmoid[0].values - Y
    #print("first", first)
    #print("X", X)
    return np.dot(first.T, X)
```

```python
def sto_grad_desc(X,Y, iters):
    lr = 0.1
    w = np.matrix(np.zeros(X.shape[1]))
    for x in range(iters):
        data = pd.concat([Y,X], axis=1)
        shuffled = data.sample(frac=1)
        shuffled_X = shuffled.iloc[0:, 1:]
        shuffled_Y = shuffled.iloc[0:, 0]
        for index in range(len(shuffled.index)):
            indivX = pd.DataFrame(shuffled_X.iloc[index])
            indivY = shuffled_Y.iloc[index]
            #print(indivX.T)
            #print(indivY)
            w = w - (lr * error_function_deriv(w, indivX.T, indivY))
    return w
sto_grad_desc(labels, classification, 10)
        
```

```python
def batch_grad_desc(X,Y,iters,batch_size):
    lr = 0.1
    w = np.matrix(np.zeros(X.shape[1]))
    for x in range(iters):
        data = pd.concat([Y,X], axis=1)
        shuffled = data.sample(frac=1, ignore_index=True)
        shuffled_X = shuffled.iloc[0:, 1:]
        shuffled_Y = shuffled.iloc[0:, 0]
        for index in range(len(shuffled.index)//batch_size+1):
            start_ind = index * batch_size
            end_ind = start_ind + batch_size
            indivX = pd.DataFrame(shuffled_X.iloc[start_ind:end_ind])
            indivY = shuffled_Y.iloc[start_ind:end_ind]
            #print(indivX)
            #print(indivY)
            w = w - (lr * error_function_deriv(w, indivX, indivY))
    return w
batch_grad_desc(labels, classification, 10, 30)
```

```python
def test_data_error(X,Y, coef):
    preds = sigmoid_func(coef, X)
    pred_val = np.where(preds >= 0.5, 1, 0)
    errors = 0
    for x in range(len(pred_val)):
        if(pred_val[x] != Y.iloc[x]):
            errors+=1
    #print(errors, 1-(errors / len(pred_val)))
    return 1-(errors / len(pred_val))
```

```python
#function to do x-fold cross validation with a given model name and features and values data
def cross_validation(folds, features, values, iters, batch_size):
    feature_cuts = []
    values_cuts = []
    interval = (len(features) // folds)+1
    maximum = len(features.index)
    for x in range(folds):
        feature_cuts.append(list(range(x*interval,min(maximum,(x+1)*interval))))
        values_cuts.append(list(range(x*interval,min(maximum,(x+1)*interval))))
    batch_errors = []
    sto_errors = []
    for x in range(folds):
        training_features = []
        training_values = []
        for y in range(folds):
            if x != y:
                training_features.extend(feature_cuts[y])
        batch_coef = batch_grad_desc(features.iloc[training_features], values.iloc[training_features], iters, batch_size)
        sto_coef = sto_grad_desc(features.iloc[training_features], values.iloc[training_features], iters)
        #print(batch_coef, sto_coef)
        batch_errors.append(test_data_error(features.iloc[feature_cuts[x]], values.iloc[feature_cuts[x]], batch_coef))
        sto_errors.append(test_data_error(features.iloc[feature_cuts[x]], values.iloc[feature_cuts[x]], sto_coef))
    print(batch_errors, sto_errors)
        
        #print(trained_model)
cross_validation(5, labels, classification, 10, 25)
```
